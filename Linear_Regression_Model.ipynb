```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9, 5))
sns.scatterplot(x='Experience Years', y='Salary', data=df)
plt.title('Scatter Plot of Experience Years vs Salary')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
x = df[['Experience Years']]
y = df['Salary']
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
year=float(input("Enter You Experience in year: "))
salary=lr.predict([[year]])
print(salary)
```
